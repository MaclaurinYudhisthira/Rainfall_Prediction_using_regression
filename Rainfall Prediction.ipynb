```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('austin_weather.csv')
```

```python
df.info()
```

```python
df.drop(columns=['Date','Events'],inplace=True)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.replace('T',0.0,inplace=True)
```

```python
df.replace('-',0.0,inplace=True)
```

```python
df.nunique()
```

```python
df=df.apply(pd.to_numeric)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(),annot=True)
```

```python
df.columns
```

```python
df.drop(columns=['TempHighF',  'TempLowF', 'DewPointHighF', 'DewPointLowF', 'HumidityHighPercent','SeaLevelPressureHighInches', 'SeaLevelPressureLowInches'],inplace=True)
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(),annot=True)
```

```python
sns.distplot(df.PrecipitationSumInches, kde=False)
```

```python
X=df.drop(columns=['PrecipitationSumInches'])
y=df.PrecipitationSumInches
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

# Train accuracy

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred=model.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

# Test accuracy

```python
y_pred=model.predict(x_train)
```

```python
mean_squared_error(y_train,y_pred)
```

```python
r2_score(y_train,y_pred)
```
